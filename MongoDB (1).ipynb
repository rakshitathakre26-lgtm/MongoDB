1.What are the key differences between SQL and NoSQL databases?

🔑 Key Differences between SQL and NoSQL Databases
1. Data Model

SQL (Relational Databases):

Data is stored in structured tables (rows and columns).

Predefined schema (structure must be known before inserting data).

Examples: MySQL, PostgreSQL, Oracle, SQL Server.

NoSQL (Non-Relational Databases):

Data is stored in various flexible formats:

Document-based (MongoDB, CouchDB)

Key-Value pairs (Redis, DynamoDB)

Column-family (Cassandra, HBase)

Graph-based (Neo4j)

Schema-less (you can store unstructured or semi-structured data).

2. Scalability

SQL:

Scales vertically (upgrade a single server with more CPU, RAM, storage).

Horizontal scaling (adding more servers) is difficult and complex.

NoSQL:

Designed for horizontal scaling (distribute data across multiple servers).

Better for large-scale applications (social media, big data, IoT).

3. Flexibility

SQL:

Rigid schema — altering tables (e.g., adding a column) can be complex.

NoSQL:

Flexible schema — easy to add new fields without affecting existing data.

4. Transactions

SQL:

Strong ACID compliance (Atomicity, Consistency, Isolation, Durability).

Ensures data integrity (e.g., banking, finance).

NoSQL:

Often BASE model (Basically Available, Soft state, Eventually consistent).

Prioritizes speed and scalability over strict consistency.

5. Query Language

SQL:

Uses Structured Query Language (SQL) (powerful and standardized).

NoSQL:

Uses various query APIs depending on database type (e.g., MongoDB uses JSON-like queries).

6. Use Cases

SQL:

Best for structured data, complex queries, transactions.

Example: Banking systems, ERP, CRM.

NoSQL:

Best for unstructured/large-scale data, fast-changing data, big data applications.

Example: Social media platforms, recommendation engines, IoT.

✅ In short:

Use SQL when you need reliability, structured data, and complex queries.

Use NoSQL when you need scalability, flexibility, and can handle unstructured or rapidly changing data.

3. Explain the concept of collections in MongoDB.

📂 Collections in MongoDB
1. What is a Collection?

A collection in MongoDB is a group of documents (similar to how a table in SQL is a group of rows).

Each collection stores related data.

Example:

A students collection might store documents like:

{ "name": "Aarav", "age": 16, "class": "10th" }
{ "name": "Sara", "age": 15, "class": "9th" }

2. How it Differs from SQL Tables
SQL (Relational DB)	MongoDB (NoSQL)
Table	Collection
Row	Document
Column	Field

👉 But unlike SQL tables, collections don’t enforce a fixed schema — documents in the same collection can have different structures.

3. Key Features

Schema-less → You can add new fields without altering the collection.

Dynamic → Two documents in the same collection can have different fields.

Indexed → Collections can have indexes to speed up queries.

4. Creating a Collection

MongoDB automatically creates a collection when you insert the first document:

db.students.insertOne({ name: "Aarav", age: 16, class: "10th" })


→ This will create a students collection if it doesn’t already exist.

Or explicitly:

db.createCollection("students")

5. When to Use Different Collections

Use separate collections for logically different data.

Example: students, teachers, courses.

But don’t over-split — use embedded documents if the data belongs together.

✅ In short:
A collection in MongoDB is like a table in SQL, but it is more flexible because it can hold documents with different fields and structures.

4.How does MongoDB ensure high availability using replication?

MongoDB ensures high availability mainly through a feature called replication. Let’s break it down step by step:

🔄 Replication in MongoDB
1. What is Replication?

Replication means keeping multiple copies of the same data on different servers.

This ensures data is not lost if one server fails.

2. Replica Set

In MongoDB, replication is achieved using a replica set.

A replica set is a group of MongoDB servers that maintain the same data set.

It typically has:

Primary Node → Handles all writes and reads (by default).

Secondary Nodes → Maintain copies of the data from the primary.

Arbiter (optional) → Participates in elections but does not store data.

3. How It Works

The primary node accepts all write operations.

Secondary nodes replicate (copy) the data from the primary using an oplog (operation log).

If the primary fails, an automatic election happens among the secondaries to choose a new primary.

Applications reconnect to the new primary, usually within a few seconds.

4. Benefits of Replication

✅ High availability → If one server goes down, another takes over.

✅ Data redundancy → Copies of data are stored across multiple servers.

✅ Disaster recovery → Even if a server crashes, data remains safe.

✅ Read scaling → You can send read queries to secondary nodes to balance the load.

5. Example Setup

A 3-node replica set:

Node 1 → Primary

Node 2 → Secondary

Node 3 → Secondary (or Arbiter)

If Node 1 (primary) fails → Node 2 becomes the new primary automatically.

✅ In summary:
MongoDB ensures high availability using replica sets where data is copied across multiple servers. If the primary server fails, a new primary is elected automatically, keeping the system up and running with minimal downtime.

6.What is the role of indexes in MongoDB, and how do they improve performance?

Indexes are one of the most important concepts for understanding how MongoDB achieves fast queries. Let’s break it down:

📑 Role of Indexes in MongoDB
1. What is an Index?

An index is a special data structure that stores a small portion of the collection’s data in a way that makes it easy to search.

Think of it like an index in a book: instead of scanning every page, you can jump straight to the page where the keyword is listed.

2. Why Indexes are Needed

Without indexes, MongoDB must perform a collection scan (check every document one by one).

With indexes, MongoDB can quickly locate documents that match the query.

Example:

db.students.find({ age: 16 })


Without an index → MongoDB checks all documents.

With an index on age → MongoDB jumps directly to matching entries.

3. How Indexes Improve Performance

Faster Query Execution → Reduces the number of documents MongoDB needs to scan.

Efficient Sorting → Queries with sort() can use indexes instead of sorting all results in memory.

Improved Read Performance → Especially helpful in large collections with millions of documents.

4. Types of Indexes in MongoDB

Single Field Index – On one field (e.g., age).

Compound Index – On multiple fields (e.g., name + age).

Multikey Index – For indexing array fields.

Text Index – For full-text search.

Geospatial Index – For location-based queries (e.g., “find cafes near me”).

5. Creating an Index
db.students.createIndex({ age: 1 })   // 1 = ascending order
db.students.createIndex({ name: -1 }) // -1 = descending order

6. Trade-offs

Pros: Speeds up queries, sorting, and searching.

Cons: Indexes take extra storage space and can slow down writes (since the index must be updated on every insert/update/delete).

✅ In summary:
Indexes in MongoDB act like a shortcut that allows the database to find and sort data much faster without scanning the whole collection. They improve read performance but come with a trade-off in storage and write speed.

7.Describe the stages of the MongoDB aggregation pipeline.

The Aggregation Pipeline in MongoDB is like an assembly line for data: documents go through a series of stages, and each stage transforms them before passing them to the next.

🔄 Stages of the MongoDB Aggregation Pipeline

Here are the main stages (most commonly used):

1. $match – Filtering Data

Filters documents (like WHERE in SQL).

Example:

{ $match: { age: { $gte: 18 } } }


→ Selects only documents where age ≥ 18.

2. $project – Shaping Data

Selects specific fields or creates new computed fields.

Example:

{ $project: { name: 1, age: 1, isAdult: { $gte: ["$age", 18] } } }


→ Keeps only name, age, and adds a new field isAdult.

3. $group – Grouping & Aggregating

Groups documents by a field and performs calculations.

Example:

{ $group: { _id: "$class", avgAge: { $avg: "$age" } } }


→ Groups by class and calculates average age.

4. $sort – Sorting Results

Sorts documents in ascending (1) or descending (-1) order.

Example:

{ $sort: { age: -1 } }


→ Sorts by age descending.

5. $limit – Limiting Output

Restricts the number of documents passed on.

Example:

{ $limit: 5 }


→ Returns only the first 5 documents.

6. $skip – Skipping Documents

Skips a given number of documents.

Example:

{ $skip: 10 }


→ Skips the first 10 results.

7. $unwind – Deconstructing Arrays

Breaks an array field into multiple documents.

Example:

{ $unwind: "$subjects" }


→ If a student has subjects: ["Math", "Science"], it becomes two separate documents.

8. $lookup – Joining Collections

Performs a left outer join between two collections.

Example:

{
  $lookup: {
    from: "teachers",
    localField: "teacher_id",
    foreignField: "_id",
    as: "teacherDetails"
  }
}


→ Joins students with teachers collection.

9. $out / $merge – Writing Results

$out → Writes results to a new collection.

$merge → Merges results into an existing collection.

✅ Example Pipeline
db.students.aggregate([
  { $match: { age: { $gte: 15 } } },         // Filter
  { $group: { _id: "$class", avgAge: { $avg: "$age" } } }, // Group
  { $sort: { avgAge: -1 } }                  // Sort
])


→ Finds the average age of students in each class, sorted by highest average.

✅ In summary:
The MongoDB aggregation pipeline processes data step by step using stages like $match, $project, $group, $sort, $limit, $unwind, $lookup, and $out. Each stage transforms the documents, making it powerful for analytics and reporting.

8.What is sharding in MongoDB? How does it differ from replication?

sharding and replication are both scaling techniques in MongoDB, but they solve different problems. Let’s unpack this clearly.

⚡ Sharding in MongoDB
1. Definition

Sharding = Horizontal Partitioning of data across multiple servers (called shards).

Each shard holds only a portion of the data.

Together, shards form one logical database.

2. Why Sharding?

Needed when a single server cannot handle large data volumes or high throughput.

Helps applications deal with Big Data and millions of requests per second.

3. How Sharding Works

Shard – A MongoDB server (or replica set) that stores part of the data.

Config Servers – Store metadata (data distribution information).

Mongos Router – Acts like a traffic cop, directing queries to the correct shard.

👉 Example:
If you shard a students collection by student_id:

Shard 1 → student_id 1–1000

Shard 2 → student_id 1001–2000

Shard 3 → student_id 2001–3000

When a query comes, mongos forwards it only to the relevant shard.

🔄 Replication in MongoDB
1. Definition

Replication = Copying the same data across multiple servers (replica set).

Provides data redundancy and high availability.

2. How It Works

One primary node (handles writes).

Multiple secondary nodes (keep copies of primary’s data).

If primary fails → automatic failover elects a new primary.

🆚 Sharding vs Replication
Feature	Replication 🌀	Sharding ⚡
Purpose	High availability, data redundancy	Handle large data + high throughput
Data Storage	All nodes store the same data	Each shard stores different data
Scaling	Vertical scaling (limited horizontal read scaling)	True horizontal scaling for big data
Use Case	Fault tolerance, backups, failover	Distributing large datasets, high-performance queries
Example	Banking system (must always be available)	Social media platform with millions of users

✅ In summary:

Replication = Copies of the same data → ensures availability & reliability.

Sharding = Splitting data across servers → ensures scalability & performance.

10.What are the ACID properties in the context of MongoDB transactions?

This ties MongoDB (a NoSQL database) back to traditional database principles. Let’s go step by step.

⚖️ ACID Properties in MongoDB Transactions

ACID stands for Atomicity, Consistency, Isolation, Durability. These are the guarantees databases provide to ensure reliable transactions. Since MongoDB 4.0, multi-document transactions support full ACID compliance.

1. Atomicity

A transaction is all or nothing.

Either all operations succeed, or none are applied.

Example: If you transfer money between accounts, both the debit and credit must happen, or neither does.

In MongoDB: Within a transaction, if an error occurs → changes are rolled back.

2. Consistency

A transaction must leave the database in a valid state.

All rules, constraints, and schemas must still hold after the transaction.

Example: If a student’s total marks must always equal the sum of subject marks, a transaction ensures this is preserved.

In MongoDB: Documents remain valid according to schema validation and rules.

3. Isolation

Transactions run as if they are the only transaction happening.

Intermediate states of a transaction are invisible to other operations.

Example: If one transaction is updating a student’s marks, another transaction won’t see half-updated data.

In MongoDB: Transactions provide snapshot isolation → other operations see data as it was before the transaction started.

4. Durability

Once a transaction is committed, the changes are permanent.

Even if the system crashes, committed data won’t be lost.

In MongoDB: Durability is ensured by writing to the journal on disk.

✅ Example in MongoDB (Python with PyMongo)
from pymongo import MongoClient

client = MongoClient("mongodb://localhost:27017/")
db = client.bank

# Start session
with client.start_session() as session:
    with session.start_transaction():
        db.accounts.update_one({"name": "Alice"}, {"$inc": {"balance": -100}}, session=session)
        db.accounts.update_one({"name": "Bob"}, {"$inc": {"balance": 100}}, session=session)


👉 If both updates succeed → transaction commits.
👉 If one fails → MongoDB rolls back, ensuring ACID properties.

✅ In summary:
MongoDB supports ACID transactions to ensure that data remains reliable, consistent, and safe even in complex, multi-document operations — making it suitable for use cases like finance, inventory management, and banking systems.

11.What is the purpose of MongoDB’s explain() function?

🔍 Purpose of explain() in MongoDB
1. What is it?

explain() provides detailed information about the execution plan of a query.

It tells you how MongoDB will or did execute the query, including which indexes are used, the number of documents scanned, and the query stages.

2. Why is it Useful?

Helps optimize queries for better performance.

Shows if a query is performing a full collection scan (slow) or using an index (fast).

Useful for debugging complex queries or aggregation pipelines.

3. Types of Execution Information

queryPlanner → Shows the plan MongoDB chooses to execute the query.

executionStats → Provides metrics like nReturned, totalKeysExamined, and totalDocsExamined.

allPlansExecution → Shows details of all query plans considered.

4. Example Usage
db.students.find({ age: 16 }).explain("executionStats")


Sample Output Highlights:

winningPlan → The plan MongoDB chose (e.g., using an index).

totalDocsExamined → How many documents were scanned.

executionTimeMillis → Time taken to execute the query.

5. Benefits

✅ Identify slow queries.

✅ Verify if indexes are being used efficiently.

✅ Improve database performance by adjusting queries or adding indexes.

✅ In short:
MongoDB’s explain() function is like a query detective — it tells you how a query runs internally, which helps debug, optimize, and tune database performance.

MongoDB is schema-less by default, but it still allows you to enforce rules on the structure of your documents using schema validation. Let’s break it down.

📐 Schema Validation in MongoDB

Schema validation lets you define rules for documents in a collection.

MongoDB checks each document before inserting or updating to ensure it meets the rules.

2. Why Use Schema Validation?

Ensures data consistency without losing the flexibility of a schema-less database.

Prevents invalid or malformed documents from being inserted.

Useful for large applications where certain fields are critical.

3. How It Works

You define validation rules using JSON Schema or MongoDB operators.

MongoDB validates documents during insert and update operations.

4. Example: Using JSON Schema
db.createCollection("students", {
   validator: {
      $jsonSchema: {
         bsonType: "object",
         required: ["name", "age", "class"],

            name: { bsonType: "string", description: "must be a string and required" },
            age: { bsonType: "int", minimum: 5, maximum: 20, description: "must be an integer between 5 and 20" },
            class: { bsonType: "string", description: "must be a string and required" }
         }
      }
   }
})


Only documents matching these rules can be inserted or updated.

Example valid document:

{ "name": "Aarav", "age": 16, "class": "10th" }


Invalid document (missing class) → rejected.

5. Validation Levels

strict → All inserts and updates must comply (default).

moderate → Only applies validation to documents that already exist or are being inserted/updated.

6. Benefits

✅ Maintain data integrity.

✅ Prevent application errors caused by unexpected data formats.

✅ Combine flexibility of NoSQL with structured data rules.

✅ In short:
MongoDB handles schema validation by letting you define rules for documents using JSON Schema or operators. This ensures your data stays consistent and valid, while still keeping the flexibility of a schema-less database.

13.What is the difference between a primary and a secondary node in a replica set?

This is fundamental to understanding replication in MongoDB. Let’s break it down clearly.

🔄 Primary vs Secondary Node in a Replica Set

A replica set is a group of MongoDB servers that maintain the same dataset to provide high availability. Nodes in a replica set can be primary or secondary.

1. Primary Node

Role: Handles all write operations (inserts, updates, deletes).

Reads: By default, most reads also go to the primary (can be configured otherwise).

Replication: Changes made on the primary are replicated to all secondary nodes.

Failover: If the primary fails, a new primary is elected automatically from the secondaries.

2. Secondary Node

Role: Maintains a copy of the primary’s data.

Replication: Continuously applies changes from the primary using the oplog (operation log).

Reads: Can serve read operations if read preference is set (default is primary).

Failover: Can be elected as the new primary if the current primary fails.

3. Key Differences
Feature	Primary Node	Secondary Node
Handles Writes	✅ Yes	❌ No
Handles Reads	✅ By default	✅ Optional (read preference)
Replicates Data	❌ Does not replicate	✅ Applies data from primary
Failover Role	Current main node	Can become primary if needed
Oplog Source	Generates oplog	Applies oplog entries from primary
4. Visualization

Imagine a primary as the “master”, and secondaries as “followers”.

All changes flow from primary → secondaries.

If the master fails, one of the followers is promoted to master automatically.

✅ In short:

Primary node = handles writes, replicates changes to secondaries.

Secondary node = maintains a copy of the data, can serve reads, and can be promoted to primary if needed.

15.Explain the concept of embedded documents and when they should be used.

Embedded documents are one of the most powerful features of MongoDB, allowing it to handle complex, hierarchical data. Let’s break it down clearly.

📦 Embedded Documents in MongoDB
1. What Are Embedded Documents?

An embedded document is a document nested inside another document.

They allow you to store related data together in a single MongoDB document.

Stored in BSON format, which supports nested structures.

Example:

{
  "name": "Aarav",
  "age": 16,
  "address": {
    "street": "123 Main St",
    "city": "Mumbai",
    "zip": "400001"
  }
}


Here, address is an embedded document inside the students document.

2. Why Use Embedded Documents?

Keep related data together → avoids joins and multiple queries.

Faster reads → data is fetched in a single query.

Natural representation → reflects real-world hierarchical relationships.

3. When to Use Embedded Documents

Use them when:

Data is closely related and frequently accessed together.

Example: Student and their address, Product and its specifications.

One-to-few relationships (small arrays or objects).

Example: A blog post with a few tags.

You want to avoid joins or multiple queries, since MongoDB doesn’t have traditional joins like SQL.

4. When Not to Use Embedded Documents

Avoid them when:

Data grows unbounded → large arrays can slow down queries.

One-to-many relationships with many items → consider referencing instead.

Example: A user with millions of posts → better to store posts in a separate collection.

5. Example: Embedded vs Referenced

Embedded:

{
  "order_id": 101,
  "customer": { "name": "Sara", "email": "sara@mail.com" },
  "items": [ {"product": "Book", "qty": 2}, {"product": "Pen", "qty": 5} ]
}


Referenced (for large datasets):

// Orders collection
{ "order_id": 101, "customer_id": 1, "items": [1001, 1002] }

// Customers collection
{ "customer_id": 1, "name": "Sara", "email": "sara@mail.com" }

// Products collection
{ "product_id": 1001, "name": "Book" }
{ "product_id": 1002, "name": "Pen" }


✅ In short:
Embedded documents let you store related data together inside a single document, improving read performance and reflecting real-world hierarchical structures. Use them for one-to-few relationships, but use references for large or unbounded relationships.

16.What is the purpose of MongoDB’s $lookup stage in aggregation?

The $lookup stage is a powerful feature in MongoDB’s aggregation framework. Let’s break it down clearly.

🔗 Purpose of $lookup in MongoDB
1. What is $lookup?

$lookup allows you to perform a left outer join between two collections.

It combines documents from a “local” collection with documents from a “foreign” collection based on a common field.

2. Why Use $lookup?

MongoDB is a NoSQL database, so joins are not native like in SQL.

$lookup lets you combine related data from multiple collections in a single aggregation pipeline.

Useful for reporting, analytics, or combining hierarchical data stored separately.

3. Basic Syntax
{
  $lookup: {
    from: "foreignCollection",   // Collection to join with
    localField: "localField",    // Field from the input documents
    foreignField: "foreignField",// Field from the foreign collection
    as: "outputArray"            // Name of the new array field to store matched documents
  }
}

4. Example

Collections:

students:

{ "_id": 1, "name": "Aarav", "class_id": 101 }
{ "_id": 2, "name": "Sara", "class_id": 102 }


classes:

{ "class_id": 101, "class_name": "10th Grade" }
{ "class_id": 102, "class_name": "9th Grade" }


Aggregation with $lookup:

db.students.aggregate([
  {
    $lookup: {
      from: "classes",
      localField: "class_id",
      foreignField: "class_id",
      as: "class_info"
    }
  }
])


Result:

{
  "_id": 1,
  "name": "Aarav",
  "class_id": 101,
  "class_info": [{ "class_id": 101, "class_name": "10th Grade" }]
}

5. Key Points

$lookup always returns an array of matching documents.

Can be used with nested pipelines for more complex joins (pipeline option in MongoDB 3.6+).

Useful for one-to-one, one-to-many relationships across collections.

✅ In short:
The $lookup stage in MongoDB joins two collections within an aggregation pipeline, allowing you to combine related data without denormalizing your database. It’s essentially MongoDB’s way of doing a SQL-like left outer join.

19. How do MongoDB transactions differ from SQL transactions?

MongoDB transactions and SQL transactions both support ACID properties, but there are important differences because MongoDB is a NoSQL, document-oriented database while SQL databases are relational. Let’s break it down.

🔄 MongoDB Transactions vs SQL Transactions
1. Scope of Transactions
Feature	SQL Transactions	MongoDB Transactions
Scope	Usually table/row-level, multi-table transactions are standard.	Can be single-document or multi-document within a replica set or sharded cluster (since v4.0).
Typical Use	Joins, multi-table updates	Multi-document updates in a collection or across collections
2. Schema

SQL: Requires a fixed schema; all tables have predefined columns.

MongoDB: Schema-less by default, but transactions can span documents with different structures.

3. Performance

SQL: Optimized for complex joins and multi-table transactions, typically faster for relational operations.

MongoDB: Single-document operations are very fast (atomic by default). Multi-document transactions are slower because they maintain ACID guarantees across multiple documents.

4. Implementation
Feature	SQL	MongoDB
ACID Guarantees	Built-in for all transactions	Single-document writes are atomic by default; multi-document transactions added in v4.0
Isolation	Usually Serializable or Repeatable Read	Snapshot isolation in multi-document transactions
Rollback	Standard	Supported within transaction; rollback all changes if any operation fails
5. Use Cases

SQL: Financial systems, accounting, ERP applications where relational integrity is key.

MongoDB: Applications with flexible schemas needing ACID across multiple documents, e.g., banking microservices, e-commerce orders, inventory updates.

6. Example Comparison

SQL Transaction:

BEGIN TRANSACTION;
UPDATE accounts SET balance = balance - 100 WHERE id = 1;
UPDATE accounts SET balance = balance + 100 WHERE id = 2;
COMMIT;


MongoDB Transaction (Python):

with client.start_session() as session:
    with session.start_transaction():
        db.accounts.update_one({"_id": 1}, {"$inc": {"balance": -100}}, session=session)
        db.accounts.update_one({"_id": 2}, {"$inc": {"balance": 100}}, session=session)


Both achieve the same atomicity, but MongoDB is designed for document-based, flexible schemas.

✅ In short:

SQL transactions are native, optimized for relational data, and work across multiple tables easily.

MongoDB transactions provide ACID across multiple documents but come with slightly higher overhead, and are optional because single-document operations are already atomic.

20.What are the main differences between capped collections and regular collections?

Capped collections are a special type of collection in MongoDB with unique properties. Let’s break down the differences clearly.

📦 Capped Collections vs Regular Collections
1. Definition

Regular Collection: A standard MongoDB collection with no size or document limits. Documents can be inserted, updated, or deleted freely.

Capped Collection: A fixed-size collection that maintains insertion order. Once it reaches its size limit, oldest documents are automatically overwritten.

2. Key Differences
Feature	Regular Collection	Capped Collection
Size Limit	No fixed limit	Fixed size specified at creation
Document Deletion	Explicit deletion needed	Automatically removes oldest documents
Insertion Order	Not guaranteed	Preserves insertion order
Updates	Full updates allowed	Cannot grow a document beyond its size
Use Case	General-purpose storage	Logs, caches, real-time data streams
Indexing	Supports all types of indexes	Only supports _id index automatically, others optional
3. Creating a Capped Collection
db.createCollection("logs", { capped: true, size: 100000, max: 1000 })


size → Maximum size in bytes.

max → Maximum number of documents (optional).

4. Use Cases for Capped Collections

Logging systems → Keep the latest N log entries.

Real-time analytics → Store only recent events for fast access.

Caching → Automatically discards old or stale data.

5. Key Advantages

Fast writes → No need to check for free space.

Predictable storage usage → Disk space is pre-allocated.

Insertion order maintained → Useful for queues or streams.

✅ In short:

Regular collections are flexible and unlimited in size.

Capped collections are fixed-size, high-performance collections that automatically overwrite old data and maintain insertion order, making them ideal for logs, caches, and streaming data.

21.What is the purpose of the $match stage in MongoDB’s aggregation pipeline?

The $match stage is one of the most commonly used stages in MongoDB’s aggregation pipeline. Let’s break it down clearly.

🔍 Purpose of $match in MongoDB Aggregation
1. What is $match?

$match is a filtering stage in the aggregation pipeline.

It selects documents from the collection that meet certain conditions, similar to a WHERE clause in SQL.

Only the documents that match the criteria move on to the next stage in the pipeline.

2. Why Use $match?

To reduce the number of documents processed by later stages.

Improves query performance, especially for large collections.

Can leverage indexes to speed up filtering when used at the beginning of the pipeline.

3. Basic Syntax
{ $match: { field: value, ... } }


Example:

db.students.aggregate([
  { $match: { age: { $gte: 15 } } }
])


Selects only students whose age is greater than or equal to 15.

4. Using Complex Conditions
db.students.aggregate([
  { $match: {
      age: { $gte: 15 },
      class: "10th Grade"
  } }
])


Only students age ≥ 15 AND in 10th Grade pass through this stage.

5. Performance Tip

Place $match as early as possible in the pipeline.

Early filtering reduces the number of documents processed in later stages like $group or $sort.

✅ In short:
The $match stage filters documents based on specified criteria, reducing the dataset for subsequent stages and improving performance and efficiency of the aggregation pipeline.

⚡ What is the WiredTiger Storage Engine?

WiredTiger is the default storage engine for MongoDB (since version 3.2).

A storage engine is the part of MongoDB that reads and writes data to disk.

WiredTiger manages how documents are stored, indexed, and retrieved efficiently.

Supports concurrent operations on different documents within a collection.

Uses multi-version concurrency control (MVCC) to avoid locking the entire collection.

Stores data and indexes in compressed form to save disk space.

Uses:

Snappy compression for fast access

zlib compression for higher compression ratios

3. Checkpointing & Journaling

Ensures data durability and crash recovery.

Periodically writes checkpoints to disk and uses a journal to recover in case of a crash.

4. Support for Large Datasets

Efficiently handles large collections and indexes.

Can store terabytes of data with predictable performance.

5. Transactions

Supports ACID multi-document transactions.

Critical for applications needing reliable, consistent writes.

✅ Why WiredTiger is Important
